6. Find the outliers and replace them outliers

```python
#using IQR
sns.boxplot(data["Shucked weight"],color='yellow')
plt.show()
```

```python
q1 = data["Shucked weight"].quantile(0.25)
q3 = data["Shucked weight"].quantile(0.75)
```

```python
IQR = q3-q1
```

```python
upper_limit = q3+1.5*IQR
lower_limit = q1-1.5*IQR
```

```python
data = data[data["Shucked weight"] < upper_limit]
```

```python
sns.boxplot(data["Shucked weight"],color='yellow')
```

```python
sns.boxplot(data['Diameter'])
```

```python
q1 = data['Diameter'].quantile(0.25)
q3 = data['Diameter'].quantile(0.75)
iqr = q3 - q1
data.loc[data['Diameter'] < q1 - 1.5*iqr, 'Diameter'] = q1
data.loc[data['Diameter'] > q3 + 1.5*iqr, 'Diameter'] = q3
sns.boxplot(x=data['Diameter'])
```

```python
sns.boxplot(x=data['Height'])
```

```python
q1 = data['Height'].quantile(0.25)
q3 = data['Height'].quantile(0.75)
iqr = q3 - q1
data.loc[data['Height'] < q1 - 1.5*iqr, 'Height'] = q1
data.loc[data['Height'] > q3 + 1.5*iqr, 'Height'] = q3
sns.boxplot(x=data['Height'])
```

# **Categorical columns and  encoding**

# **8. Split the data into dependent and independent variables.**

```python
y=data['Age']
y
```

```python
x=data.drop('Age',axis=1)
x
```

# **9. Scale the independent variables**

```python
from sklearn.preprocessing import scale
```

```python
x_scaled = pd.DataFrame(scale(x),columns = x.columns)
x_scaled.head()
```

# Train Test split

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x_scaled,y,test_size=0.3,random_state=0)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

# **Model Training**

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf = RandomForestRegressor(n_jobs = -1 , random_state = 14)
rf.fit(x_train,y_train)
```

```python
y_pred=rf.predict(x_test)
```

```python
#compare
age_pred = pd.DataFrame({'actual_y_value':y_test,'predicted_y_value':y_pred})
age_pred
```

```python
#Evaluvation
from sklearn import metrics

print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```
